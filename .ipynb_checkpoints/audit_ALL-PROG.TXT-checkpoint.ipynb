# Let us examine an ALL-PROG.TXT file output from a FANUC Oi-D controller.

### Set the filename here.

```python
filename = "sample_ALL-PROG.TXT"
```

### Read in the file.

```python
with open(filename) as f:
    dat = f.read()
```

### Remove unwanted charachters.

```python
bad_chars = [' ', '%']
for i in bad_chars : 
    dat = dat.replace(i, '') 
```

### Remove all text between parenthesis. This includes all descriptive text such as comments and alert messages.

```python
# there is no descriptive text in the sample, this is how to remove it.
import re
dat_noComments = re.sub(r" ?\([^)]+\)", "", dat)
#print(dat_noComments) # this is a lot of text...
```

### Count the remaining characters and display the result,

```python
char_count = "{:,}".format(len(dat_noComments))
print('\nTHERE ARE', char_count,  'WORKING CHARACTERS IN',   filename + "\n")
```

### Split out the program numbers and remove the duplicates.

```python
regex= "O+\d{4}"
match= re.findall(regex, dat)
program_list = list( dict.fromkeys(match) )
#print(program_list)
```

### Count the programs in the list, and display the result.

```python
program_count = len(program_list)
print('\nTHERE ARE', program_count,  'WORKING PROGRAMS IN',   filename + "\n")
```

### Count how many times each subprogram is called.

```python
program_numbers = []
for i in range(len(program_list)):
    program_numbers.append(program_list[i][1:])
#print(program_numbers)
counts = {}
for i in program_numbers:
    call = "M98P" + i
    #count how many times each program number is called
    counts[i] = dat_noComments.count(call)

    
    
#print('\n', program_count, 'SUBPROGRAMS EXIST IN THE SYSTEM\n')
#for i in counts.items():
#   print(i) 
```

### Sort the counts by frequency.

```python
from collections import OrderedDict

programs_by_frequency = OrderedDict()
for prog in sorted(counts, key=counts.get, reverse=True):
    programs_by_frequency.update({prog:counts[prog]})

#print (programs_by_frequency)
```

### Plot the results.
#### Which programs are most vital to the systems operation?
#### Are these programs protected by system parameters?
#### How are these programs being called?

```python
import numpy as np
from matplotlib import pyplot as plt 

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 30
fig_size[1] = 8
plt.rcParams["figure.figsize"] = fig_size

plt.title("Frequency of Subprogram Calls in ALL-PROG.TXT")
plt.xlabel("4-DIGIT PROGRAM NUMBER")
plt.ylabel("FREQUENCY")

x = np.arange(len(programs_by_frequency))
y = programs_by_frequency.values()

plt.bar(x, y)
plt.xticks(x + 0.5, programs_by_frequency.keys(), rotation='vertical')
plt.show()
```
